```python
import evaluate

# ğŸ”¹ Inserisci qui i path completi dei file
GROUND_TRUTH_PATH = "GT_Intervista Corvatta.txt"
PREDICTION_PATH = "audio_output/Corvatta_Intervista.txt"


def read_file(path):
    try:
        with open(path, "r", encoding="utf-8") as f:
            return f.read().strip()
    except UnicodeDecodeError:
        with open(path, "r", encoding="latin-1") as f:
            return f.read().strip()


def normalize_text(text):
    # Normalizzazione base (consigliata per ASR)
    text = text.lower()
    text = " ".join(text.split())  # rimuove spazi multipli
    return text


# ğŸ“‚ Caricamento testi
ground_truth = read_file(GROUND_TRUTH_PATH)
prediction = read_file(PREDICTION_PATH)

# ğŸ” Normalizzazione
ground_truth = normalize_text(ground_truth)
prediction = normalize_text(prediction)

# ğŸ“Š Caricamento metriche
wer_metric = evaluate.load("wer")
cer_metric = evaluate.load("cer")

# ğŸ“ˆ Calcolo metriche
wer = wer_metric.compute(predictions=[prediction], references=[ground_truth])
cer = cer_metric.compute(predictions=[prediction], references=[ground_truth])

print(f"WER: {wer:.4f} ({wer*100:.2f}%)")
print(f"CER: {cer:.4f} ({cer*100:.2f}%)")
```

```python
from jiwer import wer, cer
from collections import Counter
import re

GROUND_TRUTH_PATH = "GT_Intervista Corvatta.txt"
PREDICTION_PATH = "audio_output/Corvatta_Intervista.txt"
# =================


def read_file(path):
    with open(path, "r", encoding="latin-1") as f:
        return f.read().strip()


def normalize_text(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)  # rimuove punteggiatura
    text = " ".join(text.split())
    return text


def wer_bag_of_words(reference, hypothesis):
    ref_words = reference.split()
    hyp_words = hypothesis.split()

    ref_counter = Counter(ref_words)
    hyp_counter = Counter(hyp_words)

    # parole corrette considerando frequenze
    correct = sum(min(ref_counter[w], hyp_counter[w]) for w in ref_counter)

    total_ref = len(ref_words)

    return 1 - (correct / total_ref)


# ğŸ“‚ Caricamento testi
ground_truth = normalize_text(read_file(GROUND_TRUTH_PATH))
prediction = normalize_text(read_file(PREDICTION_PATH))

# ğŸ“Š Metriche classiche (Levenshtein-based)
wer_score = wer(ground_truth, prediction)
cer_score = cer(ground_truth, prediction)

# ğŸ“¦ Bag of Words WER
wer_bow_score = wer_bag_of_words(ground_truth, prediction)

# ğŸ“ˆ Output
print("=== Standard Metrics (Sequence-based) ===")
print(f"WER: {wer_score:.4f} ({wer_score*100:.2f}%)")
print(f"CER: {cer_score:.4f} ({cer_score*100:.2f}%)")

print("\n=== Bag of Words Metric (Order Ignored) ===")
print(f"WER BoW: {wer_bow_score:.4f} ({wer_bow_score*100:.2f}%)")
```
